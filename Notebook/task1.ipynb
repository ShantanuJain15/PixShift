```python
# Load GroundingDINO model
def load_groundingdino_model():
    config_path = "GroundingDINO_SwinT_OGC.py"
    checkpoint_path = "groundingdino_swint_ogc.pth"  # Path to pre-trained model weights
    model = load_model(config_path, checkpoint_path)
    return model

# Helper function to plot the image with bounding boxes
# def plot_image_with_boxes(image, boxes, text_labels=None):
#     plt.figure(figsize=(10, 10))
#     plt.imshow(image)

#     for i, box in enumerate(boxes):
#         box = box.cpu().numpy()
#         plt.gca().add_patch(plt.Rectangle((box[0], box[1]), box[2] - box[0], box[3] - box[1],
#                                           fill=False, color='red', linewidth=2))
#         if text_labels:
#             plt.text(box[0], box[1] - 10, text_labels[i], color='red', fontsize=12, fontweight='bold')

#     plt.axis("off")
#     plt.show()

# Main function to get bounding box from GroundingDINO
def get_groundingdino_bounding_box(image_path, text_prompt):
    # Load the GroundingDINO model
    model = load_groundingdino_model()

    print(" Load  the image")
    image_source, image = load_image(image_path)
    # print(image)
    print("process the image")
    # image_tensor = T.ToTensor()(image).unsqueeze(0)
    print("  done")

    # Make predictions using the model
    boxes, logits, phrases = predict(
        model=model,
        image=image,
        caption=text_prompt,
        box_threshold=0.3,  # Adjust this threshold for more/less boxes
        text_threshold=0.25  # Adjust this threshold based on text relevance
    )
    print("done 2")
    # Show the image with bounding boxes
    # annotated_image = annotate(image, boxes, phrases)
    annotated_frame = annotate(image_source=image_source, boxes=boxes, logits=logits, phrases=phrases)
    # plot_image_with_boxes(image, boxes, phrases)
    cv2.imwrite("annotated_image.jpg", annotated_frame)
    print("done3")
    # W, H = image.size
    # boxes = box_ops.box_cxcywh_to_xyxy(boxes) * torch.Tensor([W, H, W, H])

    return boxes, image
```

```python
def show_mask(mask, image, color=[255, 0, 0], transparency=0.9):
    mask = mask.squeeze()  # Remove extra dimensions if any
    red_mask = np.zeros_like(image)
    red_mask[:, :] = color  # Set color for the mask
    print(red_mask)
    masked_image = np.copy(image)
    masked_image[mask > 0] = ((1 - transparency) * masked_image[mask > 0] + transparency * red_mask[mask > 0]).astype(np.uint8)
    
    plt.imshow(masked_image)
    plt.axis('off')
    plt.show()
    
    
    blended_image = np.copy(image)
    mask = (mask > 0).astype(np.uint8) * 255 
    for i in range(3):  # For RGB channels
        blended_image[:, :, i] = np.where(mask > 0, 
                                          (1 - transparency) * blended_image[:, :, i] + transparency * red_mask[:, :, i],
                                          blended_image[:, :, i])
        
    
     # Displays the image
    plt.imshow(blended_image)
    plt.axis('off')
    plt.show()
    return masked_image,blended_image
```

```python
# Step 2: Use SAM for precise segmentation based on GroundingDINO's bounding box
def segment_with_sam(image_path, boxes):
    # Load SAM model
    image = upload_image(image_path)
    image_np = np.array(image)
    image_np = image_np.astype(np.float32) / 255.0
    print("done upload")
    # predictor = load_sam_model()

    # Set the image in SAM predictor
    predictor.set_image(image_np)
    print("done 4")

  
    W, H = image.size
    cx, cy, w, h = boxes[:, 0], boxes[:, 1], boxes[:, 2], boxes[:, 3]
    x_min = cx - (w / 2)
    y_min = cy - (h / 2)
    x_max = cx + (w / 2)
    y_max = cy + (h / 2)
    # Stack and multiply by image dimensions to get pixel scale
    adj_boxes = torch.stack([x_min, y_min, x_max, y_max], dim=1) * torch.tensor([W, H, W, H])
    input_box=adj_boxes[0].cpu().numpy()
    print(input_box)
    print("converted to correct_ form ")
    masks, _, _ = predictor.predict(box=input_box, point_coords=None, point_labels=None, multimask_output=False)
#     plt.figure(figsize=(10,10))
#     plt.imshow(image_np)
#     show_mask(masks, plt.gca())
    print("masked")
    masked_image, bi = show_mask(masks,image_np)
    return masks,image,masked_image,bi
```

```python
from diffusers import StableDiffusionInpaintPipeline
```

```python
# Load Stable Diffusion inpainting pipeline
pipe = StableDiffusionInpaintPipeline.from_pretrained(
    "runwayml/stable-diffusion-inpainting", 
    torch_dtype=torch.float16
).to("cuda")  # or "cpu" if you don't have a GPU
```

```python
def preprocess_image(image_path, mask_path):
    image = Image.open(image_path).convert("RGB")
    mask = Image.open(mask_path).convert("L")  # Load the mask in grayscale
    return image, mask

def inpaint_object(image, mask, prompt="clear background"):
    image = image.resize((512, 512))
    mask = mask.resize((512, 512))
    result = pipe(prompt=prompt, image=image, mask_image=mask).images[0]
    return result

def reinsert_object(inpainted_image, mask, shift_x, shift_y, prompt="reinsert object"):
    mask_np = np.array(mask)
    shifted_mask_np = np.roll(mask_np, shift=(shift_y, shift_x), axis=(0, 1))
    shifted_mask = Image.fromarray(shifted_mask_np).resize((512, 512))
    result = pipe(prompt=prompt, image=inpainted_image, mask_image=shifted_mask).images[0]
    return result
```

```python
# Example usage
image_path = "path_to_image.jpg"
mask_path = "path_to_mask.png"

image, mask = preprocess_image(image_path, mask_path)
inpainted_image = inpaint_object(image, mask, prompt="clear background")
shifted_image = reinsert_object(inpainted_image, mask, shift_x=50, shift_y=-30, prompt="reinsert object")

# Display the final image
shifted_image.show()
```
